```python
filename="Climate Guidance on Vegetation Indices.pdf"
```

```python
from PyPDF2 import PdfReader

# Load the PDF
reader = PdfReader(filename)
print(len(reader.pages))

# Extract text from each page
for page in reader.pages:
    print(page.extract_text())
    break
```

```python
import pdfplumber

# Extract tables
with pdfplumber.open(filename) as pdf:
    for page in pdf.pages:
        if page.extract_tables():
            for table in page.extract_tables():
                print(table)
```

```python
from transformers import pipeline

qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")

# Ask a question about the document
context = context
question = "What is the main purpose of the document?"
answer = qa_pipeline(question=question, context=context)
print(answer['answer'])
```

```python
question = "What are the Types of Vegetation Indices and their computation"
answer = qa_pipeline(question=question, context=context)
print(answer['answer'])
```

```python
# Initialize summarization pipeline
summarizer = pipeline("summarization")

# Summarize extracted text
text = context
summary = summarizer(text, max_length=100, min_length=25, do_sample=False)
print(summary[0]['summary_text'])
```

```python
try:
    with fitz.open(filename) as doc:
        total_pages = len(doc)
        print(f"Total pages in the document: {total_pages}")

        # Extract images from each page
        for page_number in range(total_pages):  # Ensure within range
            try:
                page = doc[page_number]
                images = page.get_images(full=True)  # Get all images
                if images:
                    print(f"Images found on Page {page_number + 1}")
                    for image_index, image in enumerate(images):
                        xref = image[0]  # Reference for the image
                        base_image = doc.extract_image(xref)
                        image_bytes = base_image["image"]
                        ext = base_image["ext"]
                        img_filename = f"page{page_number + 1}_image{image_index + 1}.{ext}"

                        # Save the image
                        with open(img_filename, "wb") as img_file:
                            img_file.write(image_bytes)
                        print(f"Saved: {img_filename}")
            except ValueError as ve:
                print(f"Error processing page {page_number + 1}: {ve}")
except Exception as e:
    print(f"Error opening the file: {e}")
```

```python
import fitz  # PyMuPDF
import pdfplumber

# filename = "example.pdf"  # Replace with your PDF file path

# Initialize containers for extracted content
extracted_text = []
extracted_images = []
extracted_tables = []

# Extract text and images using PyMuPDF
with fitz.open(filename) as doc:
    for page_number in range(len(doc)):
        page = doc[page_number]
        text = page.get_text()
        extracted_text.append(text)
        
        images = page.get_images(full=True)
        for image_index, img in enumerate(images):
            xref = img[0]
            base_image = doc.extract_image(xref)
            image_bytes = base_image["image"]
            ext = base_image["ext"]
            img_filename = f"page{page_number + 1}_image{image_index + 1}.{ext}"
            with open(img_filename, "wb") as img_file:
                img_file.write(image_bytes)
            extracted_images.append(img_filename)

# Extract tables using pdfplumber
with pdfplumber.open(filename) as pdf:
    for page in pdf.pages:
        tables = page.extract_tables()
        extracted_tables.extend(tables)
```
